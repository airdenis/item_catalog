# LESSON 4: Get Help with Your Account

## CHAPTER 1: FAQ

Before creating a support request for a question about your account or a general question about your Nanodegree program, please see our FAQ. [https://udacity.zendesk.com/hc/en-us](https://udacity.zendesk.com/hc/en-us)

You can search or browse by category for common support questions








## CHAPTER 2: Support

If you have reviewed our FAQ and still have a question, you can submit a support request from our help site. [https://udacity.zendesk.com/hc/en-us](https://udacity.zendesk.com/hc/en-us)

Please select the most relevant category for your request.



Tell us about the issue you are experiencing, and we will try to get back to you in 24 hours or less.
